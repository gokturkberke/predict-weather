```python
core_weather = weather[["PRCP","SNOW","SNWD","TMAX","TMIN"]].copy()
```

| Kısaltma | Açılımı | Anlamı |
|----------|--------|--------|
| **PRCP** | *Precipitation* | Yağış miktarı (mm veya inch) |
| **SNOW** | *Snowfall* | Kar yağışı miktarı (mm veya inch) |
| **SNWD** | *Snow Depth* | Kar kalınlığı (mm veya inch) |
| **TMAX** | *Maximum Temperature* | Günün en yüksek sıcaklığı (°C veya °F) |
| **TMIN** | *Minimum Temperature* | Günün en düşük sıcaklığı (°C veya °F) |


```python
core_weather.columns = ["precip","snow","snow_depth","temp_max","temp_min"]
```
