```python
from sklearn.linear_model import Ridge
```

```python
reg = Ridge(alpha=.1)
```

```python
predictors = ["precip","temp_max","temp_min"]
```

```python
train = core_weather.loc[:"2020-12-31"]
```

```python
test = core_weather.loc["2021-01-01":]
```

```python
reg.fit(train[predictors], train["target"]) #bagimsiz degisken - bagimli degisken
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"],predictions)
```

```python
combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
combined.columns = ["actual","predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_ #Bunlar modelin katsayıları (coefficient, β değerleri) — yani her bağımsız değişkenin hedef değişken üzerindeki etkisini gösterir.
```

```python
def create_predictions(predictors,core_weather,reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])    
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"],predictions)
    combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
    combined.columns = ["actual","predictions"]
    return error,combined
```

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
#rolling 30 “Her satır için, o satır dahil önceki 29 satırı kapsayan bir pencere oluştur” demektir. ilk 29 bu yuzden nan olur 
```

```python
core_weather
```

```python
core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]
```

```python
core_weather
```

```python
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
```

```python
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
error,combined = create_predictions(predictors,core_weather,reg)
```

```python
core_weather["monthly_avg"] = (
    core_weather["temp_max"]
    .groupby([core_weather.index.year, core_weather.index.month])
    .transform("mean")
) #Transform(): Gruplama yaptıktan sonra hesaplanan değeri orijinal DataFrame'in boyutunda geri döndürür.
```

```python
core_weather
```

```python
core_weather["day_of_year_avg"] = (
    core_weather["temp_max"]
    .groupby(core_weather.index.day_of_year)
    .transform(lambda x: x.expanding().mean())
)
```

```python
error, combined =  create_predictions(predictors,core_weather,reg)
```
