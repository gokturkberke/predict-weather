```python
# WE WILL PREDICT TOMORROW TEMP MAX
core_weather["target"] = core_weather.shift(-1)["temp_max"] #DataFrame’deki tüm satırları bir yukarı kaydırır
# ve sadece temp_max sutununu al al
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:].copy() #son satiri istemiyoruz
#son satırı silmek için satır numarasına göre işlem yapıyoruz.
#loc kullansaydık, “son satır” etiketi ne olursa olsun onu bilmemiz gerekirdi (örneğin loc["2022-01-27"]).
```

iloc = label  based / loc = integer based

```python
core_weather
```

```python
from sklearn.linear_model import Ridge
```

```python
reg = Ridge(alpha=.1)
```

```python
predictors = ["precip","temp_max","temp_min"]
```

```python
train = core_weather.loc[:"2020-12-31"]
```

```python
test = core_weather.loc[:"2021-01-01"]
```

```python
reg.fit(train[predictors], train["target"]) #bagimsiz degisken - bagimli degisken
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"],predictions)
```

```python
combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
combined.columns = ["actual","predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_ #Bunlar modelin katsayıları (coefficient, β değerleri) — yani her bağımsız değişkenin hedef değişken üzerindeki etkisini gösterir.
```
